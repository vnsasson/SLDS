```python
HR_X = hr_sampled_graph[0:len(hr_sampled_graph)-1]    # HR [0..N-1] 
print("HR_X")
print("")
print(HR_X)
print("")
print("HR_X size:", len(HR_X))
print("")

BP_X = sys_sampled_graph[0:len(sys_sampled_graph)-1]    # SYS_BP [0..N-1]
print("BP_X")
print("")
print(BP_X)
print("")
print("BP_X size:", len(BP_X))
print("")

R_X = res_sampled_graph[0:len(res_sampled_graph)-1]    # Resh h [0..N-1]
print("R_X")
print("")
print(R_X)
print("")
print("R_X size:", len(R_X))
print("")



X = np.column_stack((HR_X, BP_X, R_X)) # Mat of all signals in columns from 0 to n-1
print("Matrix X is:")
print(X)
print("")
print("")
print("")



HR_Y = hr_sampled_graph[1:len(hr_sampled_graph)] # taking the hr graph except the first sample
print(HR_Y)
print("HR_Y size:", len(HR_Y))
print()


BP_Y = sys_sampled_graph[1:len(sys_sampled_graph)] # taking the systolic bp graph except the first sample
print(BP_Y)
print("BP_Y size:", len(BP_Y))
print()


R_Y = res_sampled_graph[1:len(res_sampled_graph)] # taking the resp graph except the first sample
print(R_Y)
print("R_Y size:", len(R_Y))
print()



HR_coeff = np.dot(np.linalg.inv(np.dot(X.transpose(),X)),np.dot(X.transpose(),HR_Y))
print("a1,b1,c1 are:")
print(HR_coeff)
print()

BP_coeff = np.dot(np.linalg.inv(np.dot(X.transpose(),X)),np.dot(X.transpose(),BP_Y))
print("a2,b2,c2 are:")
print(BP_coeff)
print()

R_coeff = np.dot(np.linalg.inv(np.dot(X.transpose(),X)),np.dot(X.transpose(),R_Y))
print("a3,b3,c3 are:")
print(R_coeff)
print()


A = np.row_stack((HR_coeff, BP_coeff, R_coeff))
print("The regressed matrix is:")
print("")
print(A)
print("")
```
