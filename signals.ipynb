```python
import csv
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0, 2, 0.002)
size = len(t)
#print(size)
ecg = []
with open('sample/268269-2431-MDC_ECG_ELEC_POTL_II-500.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        ecg.append(float(row[1]))
    ecg = ecg[:size]
    plt.plot(t,ecg)
    plt.title('ECG')
    plt.xlabel('Time [sec]')
    plt.show()
```

```python
from scipy.signal import find_peaks

x = ecg
peaks, _ = find_peaks(x, height=8300)
x = np.asarray(x)
#print(peaks.shape)
#print(peaks)
plt.plot(x)
plt.plot(peaks, x[peaks], "x") ##  can't use it because its a list
#plt.show()



```
